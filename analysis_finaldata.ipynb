```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_save = './Output/'
#df = pd.read_csv(path_save + 'finaldataset_for_ML.csv', encoding='latin-1')
df = pd.read_csv(path_save + 'finaldataset_for_ML.csv', encoding='latin-1')
```

```python
display(df)
```

```python
df_anulados = df.loc[df['Anulado'] == 1, :]
df_encerrados = df.loc[df['Terminado'] == 1, :]
```

```python
display(df_anulados)
```

```python
# Count the occurrences of each value in the column
value_counts = df_anulados['ANO_EXERCICIO'].value_counts()

# Plot the counts using Matplotlib
plt.bar(value_counts.index, value_counts.values)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Projects cancelled by year')

# Annotate bars with their values
for i, v in enumerate(value_counts.values):
    plt.text(value_counts.index[i], v + 0.2, str(v), ha='center')

plt.show()
```

```python
# Count the occurrences of each value in the column
value_counts = df_encerrados['ANO_EXERCICIO'].value_counts()

# Plot the counts using Matplotlib
plt.bar(value_counts.index, value_counts.values)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Projects terminated by year')

# Annotate bars with their values
for i, v in enumerate(value_counts.values):
    plt.text(value_counts.index[i], v + 0.2, str(v), ha='center')

plt.show()
```

```python
display(df_anulados[['Parametros/Ano_Cand', 'LISTA_ANO_EXERCICIO',
 'ANO_EXERCICIO_VALIDOS',
 'ANO_EXERCICIO_VALIDOS_ORIGINAL',
 'Projeto_Valido']])
```

```python
def count_years_before(row):
    counter = 0
    candidate_year = int(row['Parametros/Ano_Cand'])
    years_list = row['LISTA_ANO_EXERCICIO'].strip('][').split(', ')
    for year in years_list:
        if int(year) < candidate_year:
            counter += 1
    return counter
```

```python
df_anulados['NR_ANOS_VALIDOS'] = df_anulados.apply(lambda x: count_years_before(x), axis=1)
```

```python
df_encerrados['NR_ANOS_VALIDOS'] = df_encerrados.apply(lambda x: count_years_before(x), axis=1)
```

```python
# Count the occurrences of each value in the column
value_counts = df_encerrados['NR_ANOS_VALIDOS'].value_counts()

# Plot the counts using Matplotlib
plt.bar(value_counts.index, value_counts.values)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Number of valid years of terminated projects')

# Annotate bars with their values
for i, v in enumerate(value_counts.values):
    plt.text(value_counts.index[i], v + 0.2, str(v), ha='center')

plt.show()
```
