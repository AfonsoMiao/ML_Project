```python
"""
Valida se conjunto de dados tem projetos para o mesmo ano
"""
def check_project_year_candidate_duplicates(df, columns=['N_Proj_anon', 'ANO_EXERCICIO']):
    new_df = df.groupby(columns).size()
    new_df = new_df.to_frame(name = 'size').reset_index()
    display(new_df[new_df['size']>1])
```

```python
def get_years(list_years, year_candidate):
    new_list = []
    for x in json.loads(list_years):
        if int(year_candidate) > int(x):
            new_list.append(x)
    return new_list
```

### Obter projetos que não têm informação anterior ao ano de candidatura

```python
df_anos_atividade = df_projetos_terminados_anulados.groupby(['N_Proj_anon', 'Parametros/Ano_Cand']).agg({'ANO_EXERCICIO': list, 'Terminado': 'first', 'Anulado': 'first'}).reset_index().copy()
#df_anos_atividade['ANO_EXERCICIO_MIN'] =  df_anos_atividade['ANO_EXERCICIO'].apply(lambda x: min(x))
df_anos_atividade.rename(columns = {'ANO_EXERCICIO':'LISTA_ANO_EXERCICIO'}, inplace = True)
df_anos_atividade
```

```python
# df_anos_atividade = df_projetos_terminados_anulados.groupby(['N_Proj_anon', 'Parametros/Ano_Cand'])['ANO_EXERCICIO'].apply(list).reset_index().copy()
# df_anos_atividade['ANO_EXERCICIO_MIN'] =  df_anos_atividade['ANO_EXERCICIO'].apply(lambda x: min(x))
# df_anos_atividade
```

```python
def exclude_greater_than(numbers_list, max_value):
    return [num for num in numbers_list if num < max_value]
```

```python
#df_anos_atividade[['LISTA_ANO_EXERCICIO', 'ANO_EXERCICIO_MIN']]
```

```python
df_anos_atividade['ANO_EXERCICIO_VALIDOS'] =  df_anos_atividade.apply(lambda x: exclude_greater_than(x['LISTA_ANO_EXERCICIO'], x['Parametros/Ano_Cand']), axis=1)
df_anos_atividade
```

```python
df_anos_atividade[df_anos_atividade['ANO_EXERCICIO_VALIDOS'].apply(lambda x: len(x) < 1)]
```

```python
len(df_anos_atividade.loc[(df_anos_atividade['ANO_EXERCICIO_VALIDOS'].apply(lambda x: len(x) < 1)) & (df_anos_atividade['Terminado'] == 1.0)])
```

```python
len(df_anos_atividade.loc[(df_anos_atividade['ANO_EXERCICIO_VALIDOS'].apply(lambda x: len(x) < 1)) & (df_anos_atividade['Anulado'] == 1.0)])
```

```python
# Filtrar projetos que não têm ANO_EXERCICIO_VALIDOS
df_anos_atividade = df_anos_atividade.loc[df_anos_atividade['ANO_EXERCICIO_VALIDOS'].apply(lambda x: len(x) >= 1)]
df_anos_atividade
```

```python
df_anos_atividade_exploded = df_anos_atividade.explode('ANO_EXERCICIO_VALIDOS')
df_anos_atividade_exploded
```
