```python
# Criar 3 colunas
# Coluna com o numero de anos antes da candidaturas

def numero_anos_atividade(row):
    lista_ano_atividade = row['LISTA_ANO_EXERCICIO'].strip('][').split(', ')
    min_ano_atividade = int(min(lista_ano_atividade))
    ano_antes_candidatura = int(row['Parametros/Ano_Cand'])
    if int(min_ano_atividade) <= int(ano_antes_candidatura):
        return ano_antes_candidatura - min_ano_atividade
    else:
        return 0
    

# Coluna flag a indicar se há informação para todos os anos, ou seja, 4 anos
def flag_info_todos_anos(row):
    lista_ano_atividade = row['LISTA_ANO_EXERCICIO'].strip('][').split(', ')
    min_ano_atividade = int(min(lista_ano_atividade))
    ano_antes_candidatura = int(row['Parametros/Ano_Cand'])
    if min_ano_atividade <= ano_antes_candidatura:
        if (ano_antes_candidatura - min_ano_atividade) == 4:
            return 1
        else:
            return 0
    else:
        return 0

# Coluna flag a indicar se há informação para algum ano
def flag_info_algum_ano(row):
    lista_ano_atividade = row['LISTA_ANO_EXERCICIO'].strip('][').split(', ')
    min_ano_atividade = int(min(lista_ano_atividade))
    ano_antes_candidatura = int(row['Parametros/Ano_Cand'])
    if min_ano_atividade <= ano_antes_candidatura:
        if (ano_antes_candidatura - min_ano_atividade) >= 0:
            return 1
        else:
            return 0
    else:
        return 0
```
